```python
   
tesla_data = yf.download("TSLA", start="2010-01-01", end="2024-12-31")
tesla_data.reset_index(inplace=True)

def make_graph(stock_data, title):
    plt.figure(figsize=(10,6))
    plt.plot(stock_data["Date"], stock_data["Close"], label="Stock Price", linewidth=2)
    plt.xlabel("Date")
    plt.ylabel("Price (USD)")
    plt.title(title)
    plt.legend()
    plt.grid(True, linestyle="--", alpha=0.6)
    plt.show()

# رسم سهم تسلا
make_graph(tesla_data, "Tesla Stock Price Trend")

```

```python

gme_data = yf.download("GME", start="2010-01-01", end="2024-12-31")
gme_data.reset_index(inplace=True)

# رسم سهم GameStop
make_graph(gme_data, "GameStop Stock Price Trend")
```
